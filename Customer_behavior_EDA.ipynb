```python
#Question 1
Which gender generates more revenue?
```

```python
df.groupby('Gender')['Purchase Amount (USD)'].sum()
```

```python
#Answer
Female customers generate slightly higher total revenue compared to male customers.
```

```python
#Question 2
Which product category is purchased most frequently?
```

```python
df['Category'].value_counts()
```

```python
#Answer:
Clothing category has the highest number of purchases.
```

```python
#Question 3
Do subscribers spend more than non-subscribers?
```

```python
df.groupby('Subscription Status')['Purchase Amount (USD)'].mean()
```

```python
#Answer:
Subscribers spend more on average than non-subscribers.
```

```python
#Question 4
Which shipping type is associated with higher spending?
```

```python
df.groupby('Shipping Type')['Purchase Amount (USD)'].mean()
```

```python
#Answer
Express shipping customers have a higher average purchase amount
```

```python
Question 5
Which age group spends the most?
```

```python
bins = [18,25,35,45,55,65,100]
labels = ['18-25','26-35','36-45','46-55','56-65','65+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df.groupby('Age Group')['Purchase Amount (USD)'].mean()
```

```python
#Answer:
Customers aged ___ spend the highest amount on average.
```
